```python
gelato_posts = []

successful_scraped_posts = []

time.sleep(3)


# Check that previosly scraped posts file exists
if os.path.exists(path + "/successful_scraped_posts.csv"):
    df_ = pd.read_csv(path + "/successful_scraped_posts.csv",
                         usecols = ['url'])
    previously_scraped_posts =  list(df_["url"])
    #print(previously_scraped_posts)
else:
    previously_scraped_posts = []


    
#alt tab to new page        
pyautogui.keyDown('alt') 
pyautogui.press('tab')     
pyautogui.keyUp('alt')
time.sleep(0.3)


# click first post
pyautogui.moveTo(687, 930)
time.sleep(1.0)
pyautogui.click()
time.sleep(5.0)


# set up checks
check_url = ""



duplicate_post = False



while duplicate_post == False:
    
    # Get url
    pyautogui.keyDown('ctrl') 
    pyautogui.press('l')     
    pyautogui.keyUp('ctrl')
    time.sleep(0.3)
    
    # Copy url
    pyautogui.keyDown('ctrl') 
    pyautogui.press('c')     
    pyautogui.keyUp('ctrl')
    time.sleep(0.3)
    
    # access clipboard
    win32clipboard.OpenClipboard()
    url = win32clipboard.GetClipboardData()
    win32clipboard.CloseClipboard()
    
    if 'https://www.instagram.com/p/CTETA4XNn2v/' in url:
        # go to next page
        pyautogui.press('right')
        time.sleep(1.0)
        
        # access clipboard
        win32clipboard.OpenClipboard()
        url = win32clipboard.GetClipboardData()
        win32clipboard.CloseClipboard()
    
    
    if url in previously_scraped_posts:
        print('already scraped', url)
        
        
        # move to main page
        pyautogui.moveTo(1415, 215)
        time.sleep(1.0)
        pyautogui.click()
        time.sleep(0.5)
        
        # go to next page
        pyautogui.press('right')
        time.sleep(1.0)
    
    else:
        # check for repeated text
        if url == check_url:
            print('break as repeated post')
            break
        else:
            check_url = url
            
        print('new url', url)
        previously_scraped_posts.append(url)

        # move to main page
        pyautogui.moveTo(1415, 215)
        time.sleep(1.0)
        pyautogui.click()
        time.sleep(0.5)


        # select everything on  page
        pyautogui.keyDown('ctrl') 
        pyautogui.press('a')     
        pyautogui.keyUp('ctrl')       
        time.sleep(0.3)

        # Copy all previoulsy selected
        pyautogui.keyDown('ctrl') 
        pyautogui.press('c')     
        pyautogui.keyUp('ctrl')
        time.sleep(0.3)

        win32clipboard.OpenClipboard()
        post_data = win32clipboard.GetClipboardData()
        win32clipboard.CloseClipboard()

        #url = ""
        text = ""
        date = ""

        try:
            # process text
            post_data1 = post_data.replace('\r',"").replace('\n', " ")
            post_data1 = post_data.split('Verified')
            text = post_data1[3].split('\r\n')


            text_out = ""
            for line in text:
                if len(line) < 2:
                    pass
                elif 'profile picture' in line:
                    break
                else:
                    #print(line)
                    text_out = text_out + line
            text_out = text_out.encode('ascii', 'ignore').decode()
            text_out


            # process date
            post_data1 = post_data.replace('\r',"").replace('\n', " ")
            post_data1 = post_data.split('Verified')
            post_data1 = post_data1[-1].split('\n\r\n\r\n')
            post_data1 = post_data1[-1].split('\r\n\r\n')
            date = post_data1[0]
        except:
            pass







        # use dict format - parse later
        post_info = {
            'url' : url,
            'text' : text_out,
            'date' : date
        }

        # append to list
        successful_scraped_posts.append(post_info)

        # go to next page
        pyautogui.press('right')
        time.sleep(8.0)

    


    

    
```

```python
# convert to df
df_successful_scraped_posts = pd.DataFrame.from_dict(successful_scraped_posts)

# if already exists, append to previous file
if os.path.exists(path + "/successful_scraped_posts.csv"):
    #with open('C:/Users/Declan/Desktop/Messina/successful_scraped_posts.csv', 'a') as outfile:
    df_successful_scraped_posts.to_csv(path + "/successful_scraped_posts.csv", mode='a', index=False, header=False)

else:
    df_successful_scraped_posts = df_successful_scraped_posts.to_csv(path + "/successful_scraped_posts.csv", index=False)
    
```
