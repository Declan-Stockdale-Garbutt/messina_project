```python
#scroll
scrolldown=driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var scrolldown=document.body.scrollHeight;return scrolldown;")
SCROLL_PAUSE_TIME = 6

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

mouse_move = 'left'

posts = []



while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    if mouse_move == 'left':
        pyautogui.moveTo(500, 100)
        mouse_move = 'right'
    else:
        pyautogui.moveTo(10, 600)
        mouse_move = 'left'
        
        
    # scrape links
    links = driver.find_elements_by_tag_name('a')
    for link in links:
    
        post = link.get_attribute('href')

        if '/p/' in post:
            posts.append(post)
            print(post)
            
    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
gelato_posts = []

previously_scraped_posts = []
successful_scraped_posts = []

#driver.get('https://www.instagram.com/gelatomessina')

try:
    #login
    time.sleep(5)
    username=driver.find_element_by_css_selector("input[name='username']")
    password=driver.find_element_by_css_selector("input[name='password']")
    username.clear()
    password.clear()
    username.send_keys("USERNAME")
    password.send_keys("PASSWORD")
    login = driver.find_element_by_css_selector("button[type='submit']").click()
except:
    pass

time.sleep(3)


# Check that previosly scraped posts file exists
if os.path.exists("C:/Users/Declan/Desktop/Messina/successful_scraped_posts.json"):
    
    # read in previous value
    with open("C:/Users/Declan/Desktop/Messina/successful_scraped_posts.json", "r") as fp:
        for line in fp:

            # remove linebreak from a current name
            # linebreak is the last character of each line
            x = line[:-1]

            # add current item to the list - read all scraped files into list
            previously_scraped_posts.append(x)

    
# create new page        
pyautogui.keyDown('ctrl') 
pyautogui.press('n')     
pyautogui.keyUp('ctrl')
time.sleep(0.3)

# write page
pyautogui.write('instagram.com') 
time.sleep(0.5)

# load page and wait to load
pyautogui.press('enter') 
time.sleep(10)
# go over all scraped posts
for post in messina_posts:
    
    # dont scrape already scraped post
    if post not in previously_scraped_posts:
        
        # might need a mouse click
        
        # create new page        
        pyautogui.keyDown('ctrl') 
        pyautogui.press('t')     
        pyautogui.keyUp('ctrl')
        time.sleep(0.3)

        # write page
        pyautogui.write(post) 
        time.sleep(0.3)
        pyautogui.press('enter')
        time.sleep(10)


        # select everything on  page
        pyautogui.keyDown('ctrl') 
        pyautogui.press('a')     
        pyautogui.keyUp('ctrl')       
        time.sleep(0.3)
        
        # Copy all previoulsy selected
        pyautogui.keyDown('ctrl') 
        pyautogui.press('c')     
        pyautogui.keyUp('ctrl')
        time.sleep(0.3)
        
        
        # Close tab
        pyautogui.keyDown('ctrl') 
        pyautogui.press('w')     
        pyautogui.keyUp('ctrl')
        time.sleep(0.3)
        
        
        # access clipboard
        win32clipboard.OpenClipboard()
        raw_post = win32clipboard.GetClipboardData()
        win32clipboard.CloseClipboard()
       
    
        text = raw_post.split('Verified')
        text = text[2].split('\r\n')
        text = text[1]
        
        
        date = raw_post.replace("\n","").replace('\r'," ")
        date = date.split('likeReply')
        date = date[-1][3:]
        date = date[:date.find('  Add a comment')]

        # use dict format - parse later
        post_info = {
            'url' : post,
            'text' : text,
            'date' : date
        }
        
        # append to list
        successful_scraped_posts.append(post_info)
        


```

```python
# convert to df
df_successful_scraped_posts = pd.DataFrame.from_dict(successful_scraped_posts)
# convert to json
df_successful_scraped_posts = df_successful_scraped_posts.to_json()

# if already exists, append to previous file
if os.path.exists("C:/Users/Declan/Desktop/Messina/successful_scraped_posts.json"):
    with open('C:/Users/Declan/Desktop/Messina/successful_scraped_posts.json', 'a') as outfile:
        outfile.write(json.dumps(df_successful_scraped_posts))
        #outfile.write(",")
        outfile.close()

else:
    with open('C:/Users/Declan/Desktop/Messina/successful_scraped_posts.json', 'w') as outfile:

        json.dump(df_successful_scraped_posts, outfile)
    

time.sleep(10.0)
```
